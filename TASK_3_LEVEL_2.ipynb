```python
#importing libraries and reading the data set

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/Dataset.csv")
```

```python
# Find restaurants with the highest and lowest number of votes
highest_votes_restaurant = df[df['Votes'] == df['Votes'].max()]
lowest_votes_restaurant = df[df['Votes'] == df['Votes'].min()]
# Calculate the correlation between Aggregate Rating and Votes
correlation = df['Aggregate rating'].corr(df['Votes'])
```

```python
print("Restaurant with the Highest Number of Votes:")
print(highest_votes_restaurant[['Restaurant Name', 'Aggregate rating', 'Votes']])
print("\nRestaurant with the Lowest Number of Votes:")
print(lowest_votes_restaurant[['Restaurant Name', 'Aggregate rating', 'Votes']])
print("\nCorrelation between Aggregate Rating and Votes:", correlation)
```

```python
# Create a scatter plot to visualize the correlation
plt.figure(figsize=(12,6))
sns.scatterplot(x=df['Votes'], y=df['Aggregate rating'])
plt.title('Correlation between Votes and Aggregate rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.show()
```
